```python
from tensorflow import keras 
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
```

```python
import numpy as pandas
```

```python
import cv2
from sklearn.linear_model import LinearRegression

```

```python
(x_train , y_train), (x_test , y_test) = keras.datasets.cifar10.load_data()
```

```python
x_train.shape,x_test.shape
```

maintain the pixels of all the image we use normalization
 

```python
# Normalization
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
# sklearn uses 2d array as input 
nsamples, nx, ny, nrgb = x_train.shape
x_train2 = x_train.reshape((nsamples,nx*ny*nrgb))
```

```python
nsamples, nx, ny, nrgb = x_test.shape
x_test2 = x_test.reshape((nsamples,nx*ny*nrgb))
```

```python
model=RandomForestClassifier()
```

```python
model.fit(x_train2,y_train)
```

```python
y_pred=model.predict(x_test2)
y_pred
```

```python
accuracy_score(y_pred,y_test)
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```

```python
# here we can see by Randomforest classifier accuracy is 47%
```

NOW IMPLEMENTING KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=7)
```

```python
knn.fit(x_train2,y_train)
```

```python
y_pred_knn=knn.predict(x_test2)
y_pred_knn
```

```python
accuracy_score(y_pred_knn,y_test)
print(classification_report(y_pred_knn,y_test))
```

```python
confusion_matrix(y_pred_knn,y_test)
```

```python
#KNN Classifier shows only 34% accuracy 
```
